**merging the all five years data for training**

```python
concat = pd.concat([df2, df3, df4, df5,df6],ignore_index=True)
concat
```

**considering only opening data values for analyzing and prediction purpose**

```python
train_data=concat.iloc[:,2]
train_data
```

```python
test=onemonth_data.iloc[:,2]
test_data=test.reset_index(drop=True)
test_data
```

#### cleaning the data

```python
train_data = train_data.apply(lambda x: float(x.replace(',', ''))).values
train_data = np.array(train_data).reshape(-1, 1)
```

```python
test_data = test_data.apply(lambda x: float(x.replace(',', ''))).values
test_data
```

```python
##feature scaling
from sklearn.preprocessing import MinMaxScaler
sc=MinMaxScaler()
train_data_scaled=sc.fit_transform(train_data)
```

```python
train_data_scaled.shape
```

```python
x_train=[]
y_train=[]
for i in range(60,1252):
    x_train.append(train_data_scaled[i-60:i,0])
    y_train.append(train_data_scaled[i,0])
    
x_train,y_train=np.array(x_train),np.array(y_train)
```

```python
x_train=np.reshape(x_train,(x_train.shape[0],x_train.shape[1],1))
```

### importing keras libraries

```python

from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
```

```python
regressor=Sequential()
```

### Adding layers

```python
regressor.add(LSTM(units=50,return_sequences=True,input_shape=(x_train.shape[1],1)))
regressor.add(Dropout(0.2))
```

```python
regressor.add(LSTM(units=50,return_sequences=True))
regressor.add(Dropout(0.2))
```

```python
regressor.add(LSTM(units=50,return_sequences=True))
regressor.add(Dropout(0.2))
```

```python
regressor.add(LSTM(units=50))
regressor.add(Dropout(0.2))
```

```python
regressor.add(Dense(units=1))
```

```python
regressor.compile(optimizer='adam',loss='mean_squared_error')
```

### Training the data

```python
regressor.fit(x_train,y_train,epochs=100,batch_size=32)
```

```python
pred=regressor.predict(x_train)
```

```python
import math

from sklearn.metrics import mean_squared_error
rmse=math.sqrt(mean_squared_error(y_train,pred))
rmse
```

```python
real_stock_price = np.array(test_data).reshape(-1, 1)
```

```python
dataset_total=pd.concat((concat['OPEN '],onemonth_data['OPEN ']),axis=0)
dataset_total = dataset_total.apply(lambda x: float(x.replace(',', '')))

inputs= dataset_total[len(dataset_total)-len(onemonth_data):].values
inputs = np.array(dataset_total).reshape(-1, 1)
inputs= sc.transform(inputs)
x_test=[]
for i in range(60,80):
    x_test.append(inputs[i-60:i,0])
x_test=np.array(x_test)
x_test=np.reshape(x_test,(x_test.shape[0],x_test.shape[1],1))
predicted_stock_price=regressor.predict(x_test)
predicted_stock_price=sc.inverse_transform(predicted_stock_price)
```

##  visualising the results

```python

plt.plot(real_stock_price,color='red',label="Real tcs Share Price")
plt.plot(predicted_stock_price,color='blue',label='Predicted tcs share price')
plt.title('TCS stock price prediction')
plt.xlabel('Time')
plt.ylabel('TCS stock price')
plt.legend()
plt.show()
```
